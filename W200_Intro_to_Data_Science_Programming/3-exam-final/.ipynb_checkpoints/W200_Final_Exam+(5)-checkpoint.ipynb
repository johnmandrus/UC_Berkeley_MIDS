a) The following class ```Cart``` and method ```add_to_cart``` are parts of a larger program used by a mobile phone company.  The method ```add_to_cart``` will work when an object of type MobileDevice or of type ServiceContract is passed to it. State whether the method ```add_to_cart``` is a demonstration of the following items (yes/no) and the reasoning (1-2 sentences): 

    1. Inheritance
    2. Polymorphism
    3. Duck typing
    4. Top-down design
    5. Functional programming

````python
# Method:
class Cart():
    
    def __init__(self):
        self.cart = []
        self.total = 0
    
    def add_to_cart(item):
        self.cart.append(item)
        self.total += item.price
````

- a) Your answer here


1. Inheritance: NO

The Cart() class is not defined in a way that would indicate it is a child class and the add_to_cart function does not inherit functionality from a parent class.
2. Polymorphism: YES

The add_to_cart method accepts multiple types of objects and can adjust the price and contents of the cart regardless of input type.
3. Duck typing: YES

The add_to_cart method accepts multiple types of objects and references a price attribute regardless of input type. So long as the input is a MobileDevice or ServiceContract, and each of these classes has a numberical price attribute, the method will function properly.
4. Top-down design: NO

The add_to_cart method does not reference a function that has yet to be implemented and is therefore not an example of top down design.
5. Functional programming: NO

The function add_to_cart modifies the self.cart and self.total variables, which are outside of the scope of the function. In functional programming, functions do not modify variables outside of their scope.

## 3: Bitcoin coding problem (20 points):

#### Bitcoin

Consider a record of a one-time investment in bitcoin with value of that investment tracked monthly, provided as an (ordered) tuple of dictionaries, where each dictionary comprises one key for the month and corresponding one value for the value of the investment, and the first entry (Jan 2018) is the initial investment, like in `data` below.

Write Python code to take such a record of any length (the below <i>data</i> is only a sample), and output a table/dataframe comprising a row for each month with columns for date, start balance, and return. **Print out this table/dataframe.**

Also, visualize the record as two vertically arranged plots. 
- The top plot should show a line plot of start balance vs. month
- The bottom plot should show a bar plot of return vs. month, with a black horizontal line at return=0, and bars color-coded such that positive returns are green and negative returns are red.  
- The two plots' horizontal axes should align. Demonstrate that your code works by applying it to `data`.

Notes: <br>
- The gain for each period is the end balance minus the start balance. 
- The growth factor for each period is the end balance divided by the start balance.  
- The return for each period is the growth factor minus 1.

````python
data = ({"Jan 2018":1000},{"Feb 2018":1100},{"Mar 2018":1400},{"Apr 2018":700},{"May 2018":800},{"Jun 2018":500})
data
````

````python
# 3) Your code here
import matplotlib.pyplot as plt
import pandas as pd

dat = []
for i in data:
    month = list(i.keys())
    price = list(i.items())
    dat.append([month[0], price[0][1]])

df = pd.DataFrame(dat,columns=['month','price'])
df['growth'] = df['price'].diff()
df['growth'][0] = 0
df['growth'] = df['growth']/df['price']
print(df)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.set_ylabel('Price')
ax1.set_title('Value of Bitcoin Investment')
ax1.plot(df['month'], df['price'])

ax2.set_xlabel('Month')
ax2.set_ylabel('Return')
ax2.set_title('Monthly Return')
ax2.bar(df['month'], df['growth'], color=(df['growth'] > 0).map({True: 'g', False: 'r'}))
ax2.axhline(y=0, color='k', linestyle='-')

fig.tight_layout()

plt.show()
````

### VCF file description (Summarized from version 4.1)

```
* The VCF specification:

VCF is a text file format which contains meta-information lines, a header line, and then data lines each containing information about a position in the genome. The format also can contain genotype information on samples for each position.

* Fixed fields:

There are 8 fixed fields per record. All data lines are **tab-delimited**. In all cases, missing values are specified with a dot (‘.’). 

1. CHROM - chromosome number
2. POS - position DNA nuceleotide count (bases) along the chromosome
3. ID - The unique identifier for each mutation
4. REF - reference base(s)
5. ALT - alternate base(s)
6. FILTER - filter status
7. QUAL - quality
8. INFO - a semicolon-separated series of keys with values in the format: <key>=<data>

```
### Applicable INFO field specifications

```
GENEINFO = <Gene name>
CLNSIG =  <Clinical significance>
CLNDN = <Disease name>
```

````python
#TASK:         Determine which gene mutations are considered dangerous
#DELIVERABLE:  DataFrame of first 100 harmful mutations, missing values replaced with "Not_Given"

#How do you define harmful mutations?
#    (a) For the purpose of this question, I will only identify mutations that are listed as pathogenic as harmful 

#Imports
import csv
import numpy as np

#Use csv module to import tab delimited txt file as list of lists
data = list(csv.reader(open('clinvar_final.txt', "rt"), delimiter = '\t'))
clean_data = []

#Remove header rows and other junk rows
print('Full Data Length: ', len(data))
for i in data:
    if len(i) == 8:
        clean_data.append(i)
print('Length of Data minus header rows: ', len(clean_data))

#Convert to numpy matrix
data = np.array(clean_data)
data = data[1:, 0:7]
data_df = pd.DataFrame(data, columns=['CHROM', 'POS', 'ID', 'REF', 'ALT', 'FILTER', 'QUAL'])

print('data_df complete')

#Break up info column into useful values
info = []
for i in clean_data:
    info.append(i[7].split(';'))

clean_info = []
for i in info:
    row = {'GENEINFO':'Not_Given', 'CLINSIG':'Not_Given', 'CLNDN':'Not_Given'}
    for j in i:
        if 'GENEINFO' in j:
            row['GENEINFO'] = j[9:] 
        if 'CLNSIG' in j:
            row['CLINSIG'] = j[7:]
        if 'CLNDN' in j:
            row['CLNDN'] = j[6:]
    clean_info.append([row['GENEINFO'], row['CLINSIG'], row['CLNDN']])
    
info_df = pd.DataFrame(clean_info[1:], columns=['GENEINFO', 'CLINSIG', 'CLNDN'])
print('info_df complete')

df = pd.concat([data_df, info_df],axis=1,ignore_index=True)
df.columns = ['CHROM', 'POS', 'ID', 'REF', 'ALT', 'FILTER', 'QUAL', 'GENEINFO', 'CLINSIG', 'CLNDN']
print('df concatenation complete')
df[0:10]



````
